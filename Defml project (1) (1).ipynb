```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,9))
plt.title('Age wise Analysis')
plt.xlabel("Age Group")
plt.ylabel("Smoke_cig_currently")
plt.bar(df.Age,df.Smoke_cig_currently)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(9,9))
sns.barplot(x='Gender',y='Smoke_cig_currently',data=df,palette="Blues")
plt.title('Gender wise Analysis')
plt.xlabel("Gender")
plt.ylabel("Smoke_cig_currently")
plt.show()
```

```python
sns.barplot(x='Age',y='Smoke_cig_currently',hue='depression level',data=df,palette="Blues")
plt.title('Age wise Analysis')
plt.xlabel("Age")
plt.ylabel("Smoke_cig_currently")
plt.show()
```
